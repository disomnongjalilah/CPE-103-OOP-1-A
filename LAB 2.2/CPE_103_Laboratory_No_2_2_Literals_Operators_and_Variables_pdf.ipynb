```python
class StudentGrade:
    def __init__(self, name):
        self.name = name
        self.prelim_exam_score = 0
        self.midterm_exam_score = 0
        self.final_exam_score = 0
        self.activity_scores = []
        self.quiz_scores = []
        self.assignment_scores = []

    def input_scores(self):
        print("\nPrelim Grade")
        # Prelim Exam Score
        self.prelim_exam_score = float(input("Enter prelim score exam: "))

        # Activities, Quizzes, and Assignments for Prelim
        self.activity_scores = [
            float(input("Activity No.1 Score: ")),
            float(input("Activity No.2 Score: ")),
            float(input("Activity No.3 Score: "))
        ]
        self.quiz_scores = [
            float(input("Quiz No.1 Score: ")),
            float(input("Quiz No.2 Score: "))
        ]
        self.assignment_scores = [
            float(input("Assignment No. 1 Score: ")),
            float(input("Assignment No. 2 Score: "))
        ]

        print("\nMidterm Grade")
        # Midterm Exam Score
        self.midterm_exam_score = float(input("Enter midterm score exam: "))

        # Activities, Quizzes, and Assignments for Midterm
        self.activity_scores += [
            float(input("Activity No.1 Score: ")),
            float(input("Activity No.2 Score: ")),
            float(input("Activity No.3 Score: "))
        ]
        self.quiz_scores += [
            float(input("Quiz No.1 Score: ")),
            float(input("Quiz No.2 Score: "))
        ]
        self.assignment_scores += [
            float(input("Assignment No. 1 Score: ")),
            float(input("Assignment No. 2 Score: "))
        ]

        print("\nFinal Grade")
        # Final Exam Score
        self.final_exam_score = float(input("Enter final score exam: "))

        # Activities, Quizzes, and Assignments for Final
        self.activity_scores += [
            float(input("Activity No.1 Score: ")),
            float(input("Activity No.2 Score: ")),
            float(input("Activity No.3 Score: "))
        ]
        self.quiz_scores += [
            float(input("Quiz No.1 Score: ")),
            float(input("Quiz No.2 Score: "))
        ]
        self.assignment_scores += [
            float(input("Assignment No. 1 Score: ")),
            float(input("Assignment No. 2 Score: "))
        ]

    def calculate_grades(self):
        # Prelim Grade Calculation
        prelim_activity_avg = sum(self.activity_scores[:3]) / 3
        prelim_quiz_avg = sum(self.quiz_scores[:2]) / 2
        prelim_assignment_avg = sum(self.assignment_scores[:2]) / 2

        prelim_cs = (prelim_activity_avg * 0.5) + (prelim_quiz_avg * 0.3) + (prelim_assignment_avg * 0.2)
        prelim = self.prelim_exam_score * 0.5
        self.prelim_grade = prelim + (prelim_cs * 0.5)

        # Midterm Grade Calculation
        midterm_activity_avg = sum(self.activity_scores[3:6]) / 3
        midterm_quiz_avg = sum(self.quiz_scores[2:4]) / 2
        midterm_assignment_avg = sum(self.assignment_scores[2:4]) / 2

        midterm_cs = (midterm_activity_avg * 0.5) + (midterm_quiz_avg * 0.3) + (midterm_assignment_avg * 0.2)
        midterm = self.midterm_exam_score
        self.midterm_grade = (midterm_cs * 0.5) + (midterm * 0.5)

        # Final Grade Calculation
        final_activity_avg = sum(self.activity_scores[6:9]) / 3
        final_quiz_avg = sum(self.quiz_scores[4:6]) / 2
        final_assignment_avg = sum(self.assignment_scores[4:6]) / 2

        final_cs = (final_activity_avg * 0.5) + (final_quiz_avg * 0.3) + (final_assignment_avg * 0.2)
        final = self.final_exam_score
        self.final_grade = (final_cs * 0.5) + (final * 0.5)

        # Midterm and Final Grades Calculation
        midterm121 = (self.prelim_grade * 0.3) + (self.midterm_grade * 0.7)
        self.final_grade_121 = (midterm121 * 0.3) + (self.final_grade * 0.7)

    def assign_letter_grade(self):
        # Letter Grade Assignment based on final grade
        if self.final_grade_121 >= 98:
            return "Grade: 1.0"
        elif self.final_grade_121 >= 95:
            return "Grade: 1.25"
        elif self.final_grade_121 >= 92:
            return "Grade: 1.5"
        elif self.final_grade_121 >= 90:
            return "Grade: 1.75"
        elif self.final_grade_121 >= 86:
            return "Grade: 2.0"
        elif self.final_grade_121 >= 85:
            return "Grade: 2.25"
        elif self.final_grade_121 >= 80:
            return "Grade: 2.50"
        elif self.final_grade_121 <= 60:
            return "Grade: 3.0"
        else:
            return "Grade: 5.0"

    def display_results(self):
        print(f"\nPrelim Grade: {self.prelim_grade}")
        print(f"Midterm Grade: {self.midterm_grade}")
        print(f"Final Grade: {self.final_grade_121}")
        print(self.assign_letter_grade())

# Main execution
if __name__ == "__main__":
    name = input("\nEnter your name: ")
    student = StudentGrade(name)

    student.input_scores()
    student.calculate_grades()
    student.display_results()
```
